```python
%%time
pts = linspace(0.01,1.5,100)
rat1h=zeros(len(pts))
rat2h=zeros(len(pts))
fr.g_k = 'log_b'
gr.g_k = 'log_b_lgm'
hr.g_k = 'll_lgm_n'

i=0
zs = [0.25,0.35,0.45,0.6]
#zh = 0.35
mss = 1.5
for zh in zs:
    rat1h=zeros(len(pts))
    rat2h=zeros(len(pts))
    rat3h=zeros(len(pts))

    i=0
    for pt in pts:

        rat1h[i]=fr.ratio('lbd',zh,pt,f_prm,pp,mss)
        rat2h[i]=gr.ratio('lbd',zh,pt,f_prm,pp,mss)
        rat3h[i]=hr.ratio('lbd',zh,pt,f_prm,pp,mss)
        #if rat1h[i]<rat1h[i-1]:print(i)
        i+=1
    plt.suptitle('zh = '+str(zh))
    plot(pts,rat1h,label=str(fr.bmax)+'log_b')
    plot(pts,rat2h,label='bmax'+str(gr.bmax)+'log_b_lgm')
    plot(pts,rat3h,label='bmax'+str(gr.bmax)+'ll_lgm_n')

    legend()
    show() 
```
