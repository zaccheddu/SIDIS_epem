```python
pts = linspace(0.08,1.1,500)
rat1h=zeros(len(pts))
rat2h=zeros(len(pts))
fr.g_k = 'exp_lgm'
gr.g_k='new_lgm'
hr.g_k = 'log_b'
jr.g_k = 'blny'
i=0
zs = [0.25,0.35,0.45,0.6]
#zh = 0.35
mss = 1.
for zh in zs:
    rat1h=zeros(len(pts))
    rat2h=zeros(len(pts))
    rat3h=zeros(len(pts)) 
    rat4h=zeros(len(pts))
    i=0
    for pt in pts:

        rat1h[i]=fr.ratio('lbd',zh,pt,f_prm,wdt_pol,mss)
        rat2h[i]=gr.ratio('lbd',zh,pt,f_prm,wdt_pol,mss)
        rat3h[i]=hr.ratio('lbd',zh,pt,f_prm,wdt_pol,mss)
        rat4h[i]=jr.ratio('lbd',zh,pt,f_prm,wdt_pol,mss)
        i+=1
    plt.suptitle('zh = '+str(zh))
    plot(pts,rat1h,label='exp')
    plot(pts,rat2h,label='quadr')
    plot(pts,rat3h,label='logs')
    plot(pts,rat4h,label='old')
    legend()
    show() 
```
