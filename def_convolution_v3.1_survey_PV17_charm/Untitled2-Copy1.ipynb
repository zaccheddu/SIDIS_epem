```python
unpol=array([])
for h1,h2,zz1,zz2 in zip(dati_lh2.h1,dati_lh2.h2,dati_lh2.z1,dati_lh2.z2):

    if h1== 300: had1='lbd'
    elif h1== 310: had1='lbd_b'
    if h2== 100: had2='pi+'
    elif h2== 105: had2='pi-'
    elif h2== 200: had2='k+'
    elif h2==  205: had2='k-'
    unpol=np.append(unpol,func.denominator(had1,had2,zz1,zz2))
    #print(unpol)
```
