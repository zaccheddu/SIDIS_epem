```python
zlbd=[.25,.35,.45,.6]
zpks=linspace(0.22,0.75,30)
```

```python
fnc = polarization(0.27)
fnc.scale = 10.58
fnc.mass = 1.115 
fnc.frag2 = 'dss'
fnc.mdl_num = 'gauss' 
fnc.mdl_den = 'pwr_lw_star' 

fnc.g_k = 'PV17' 
fnc.charm='no'
fnc.nf=3

fnc2 = polarization(0.27)
fnc2.scale = 10.58
fnc2.mass = 1.115 
fnc2.frag2 = 'dss'
fnc2.mdl_num = 'gauss' 
fnc2.mdl_den = 'pwr_lw_star' 

fnc2.g_k = 'PV17' 
fnc2.charm='yes'
fnc2.nf=4
```

```python
lim=[-0.00005,0.07]
fit_l=4.  # dimensione linea fit
y_lbl=50
x_lbl =50
fig, axes = plt.subplots(1,4)
ct=1
for z1,ax in zip(zlbd,axes):
    den=array([])
    den2=array([])
    
    for z2 in zpks:
        den=np.append(den,fnc.denominator('lbd','pi+',z1,z2))
        den2=np.append(den2,fnc2.denominator('lbd','pi+',z1,z2))

    ax=plt.subplot(1,len(zlbd),ct)
    ax.plot(zpks,den,linewidth=fit_l,label='den w/o charm')
    ax.plot(zpks,den2,linewidth=fit_l,label='den w/ charm')
    ax.plot(zpks,den2-den,linewidth=fit_l,label='charm')

    ax.set_ylim(lim)
    ax.tick_params(axis='both', which='major', labelsize=28)
    if ct >1: ax.set_yticklabels([])
    if ct ==4 :legend(loc='upper right',bbox_to_anchor=(1.,0.9), fontsize=25,frameon=False), 
    if ct==1:ylabel('denominatore',size=y_lbl) 
    fig.text(0.45, 0.7, r'$\Lambda$ + $\pi^+$', ha='center',size=x_lbl)


    ct+=1
fig.subplots_adjust(top=0.8,bottom=0.2,left=0.105,right=0.99,hspace=0.2,wspace=0.0)
fig.set_size_inches(19.5, 10.5, forward=True)
fig.savefig('denominatore_Lb_pi+.pdf')
```
