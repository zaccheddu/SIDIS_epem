```python
from first_moments import*
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import*
from matplotlib.pyplot import*
```

```python
dbl_mod=pd.read_csv('first_moments_2h_3scenario.csv')
dbl_mod
```

```python
y_lbl=45
x_lbl =45

lnwd = 3.5


title_s=38 #dimensione titolo bin

fit_l=2.  # dimensione linea fit

t_sz=40
fnt_s=40
```

```python

dbl_mod['UPDO'] = dbl_mod['UP'] +dbl_mod['DO'] 
dbl_mod['UPDO_max'] = dbl_mod['UP_max'] +dbl_mod['DO_max'] 
dbl_mod['UPDO_min'] = dbl_mod['UP_min'] +dbl_mod['DO_min'] 
```

```python
fig1, ax = plt.subplots()
#
fill_between(dbl_mod.zs,dbl_mod.UP_min,dbl_mod.UP_max,alpha=0.5,facecolor='red')
plot(dbl_mod.zs,dbl_mod.UP,label='(3) up',color='red',linestyle='--',linewidth=lnwd)
#
fill_between(dbl_mod.zs,dbl_mod.DO_min,dbl_mod.DO_max,alpha=0.5,facecolor='cornflowerblue')
plot(dbl_mod.zs,dbl_mod.DO,label='(3) down',color='blue',linestyle='--',linewidth=lnwd)
plt.ylabel(r'$ z\; D^{\bot (1)}_{1T} (z)$',size=y_lbl)

plt.ylim([-0.0002,None] )
plt.xlim([0.185,0.7] )
plt.tick_params(axis='both', which='major', labelsize=t_sz)
#tick_params(axis='x', which='both', bottom=False,top=False,labelbottom=False)
#plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
#plt.title("up ",fontsize=title_s,x=0.46, y=0.8)
#ax.yaxis.set_major_locator(MultipleLocator(0.005))
#ax.yaxis.set_minor_locator(MultipleLocator(0.001))

ax.xaxis.set_minor_locator(MultipleLocator(0.02))
plt.xlabel(r'$z$',size=x_lbl)
#grid(True,lw=1.)
plt.legend(loc='upper right', fontsize=fnt_s,frameon=False)

#plt.text(0.3, 0.006 , r'(a)', fontsize=45)
plt.text(0.47, 0.004 , r'$Q=10.58$ GeV', fontsize=40)
fig1.subplots_adjust(top=0.9,bottom=0.2,left=0.230,right=0.76,hspace=0.,wspace=0.)

fig1.set_size_inches(20., 10., forward=True)
fig1.savefig('firsm_updo_split3.pdf')
fig1.savefig('firsm_updo_split3.png')

#show()
#fig1.show()

```

```python
fig1, ax = plt.subplots()
#
fill_between(dbl_mod.zs,dbl_mod.UPDO_min,dbl_mod.UPDO_max,alpha=0.5,facecolor='red')
plot(dbl_mod.zs,dbl_mod.UPDO,label='(3)',color='red',linestyle='--',linewidth=lnwd)
#
plt.ylabel(r'$ z\; D^{\bot (1)}_{1T} (z)$',size=y_lbl)

plt.ylim([-0.0002,None] )
plt.xlim([0.185,0.7] )
plt.tick_params(axis='both', which='major', labelsize=t_sz)
#tick_params(axis='x', which='both', bottom=False,top=False,labelbottom=False)
#plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.title("up + down",fontsize=title_s,x=0.46, y=0.8)
#ax.yaxis.set_major_locator(MultipleLocator(0.005))
#ax.yaxis.set_minor_locator(MultipleLocator(0.001))

ax.xaxis.set_minor_locator(MultipleLocator(0.02))
plt.xlabel(r'$z$',size=x_lbl)
#grid(True,lw=1.)
plt.legend(loc='upper right', fontsize=fnt_s,frameon=False)

plt.text(0.3, 0.015 , r'(a)', fontsize=45)
plt.text(0.47, 0.004 , r'$Q=10.58$ GeV', fontsize=40)
fig1.subplots_adjust(top=0.9,bottom=0.2,left=0.230,right=0.76,hspace=0.,wspace=0.)

fig1.set_size_inches(20., 10., forward=True)
fig1.savefig('firsm_updo3.pdf')
#show()
#fig1.show()

```

```python

dbl_mod['UPDOb'] = dbl_mod['UPb'] +dbl_mod['DOb'] 
dbl_mod['UPDOb_max'] = dbl_mod['UPb_max'] +dbl_mod['DOb_max'] 
dbl_mod['UPDOb_min'] = dbl_mod['UPb_min'] +dbl_mod['DOb_min'] 
```

```python
fig1, ax = plt.subplots()
#
fill_between(dbl_mod.zs,dbl_mod.UPb_min,dbl_mod.UPb_max,alpha=0.5,facecolor='red')
plot(dbl_mod.zs,dbl_mod.UPb,label=r'(3) $\overline{ \rm up}$',color='red',linestyle='--',linewidth=lnwd)
#
fill_between(dbl_mod.zs,dbl_mod.DOb_min,dbl_mod.DOb_max,alpha=0.5,facecolor='cornflowerblue')
plot(dbl_mod.zs,dbl_mod.DOb,label=r'(3) $\overline{ \rm down}$  ',color='blue',linestyle='--',linewidth=lnwd)
plt.ylabel(r'$ z\; D^{\bot (1)}_{1T} (z)$',size=y_lbl)

#plt.ylim([-0.0002,None] )
plt.xlim([0.185,0.7] )
plt.tick_params(axis='both', which='major', labelsize=t_sz)
#tick_params(axis='x', which='both', bottom=False,top=False,labelbottom=False)
#plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
#plt.title("up ",fontsize=title_s,x=0.46, y=0.8)
#ax.yaxis.set_major_locator(MultipleLocator(0.005))
#ax.yaxis.set_minor_locator(MultipleLocator(0.001))

ax.xaxis.set_minor_locator(MultipleLocator(0.02))
plt.xlabel(r'$z$',size=x_lbl)
#grid(True,lw=1.)
plt.legend(loc='lower right', fontsize=fnt_s,frameon=False)

plt.text(0.3, -0.012 , r'(b)', fontsize=45)
plt.text(0.47, -0.01 , r'$Q=10.58$ GeV', fontsize=40)
fig1.subplots_adjust(top=0.9,bottom=0.2,left=0.230,right=0.76,hspace=0.,wspace=0.)

fig1.set_size_inches(20., 10., forward=True)
fig1.savefig('firsm_upbdob_split3.pdf')
fig1.savefig('firsm_upbdob_split3.png')

#show()
#fig1.show()

```
