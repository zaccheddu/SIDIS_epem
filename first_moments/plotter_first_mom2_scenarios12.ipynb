```python
from first_moments import*
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import*
from matplotlib.pyplot import*
```

```python
dbl_mod=pd.read_csv('first_moments_2h_1scenario.csv')
a =pd.read_csv('first_moments_2h_2scenario.csv')
```

```python
y_lbl=45
x_lbl =45

lnwd = 3.5


title_s=38 #dimensione titolo bin

fit_l=2.  # dimensione linea fit

t_sz=40
fnt_s=40
```

```python
figure(figsize=(15, 10))
fill_between(dbl_mod.zs,dbl_mod.UP_min,dbl_mod.UP_max,alpha=0.5,facecolor='red')
plot(dbl_mod.zs,dbl_mod.UP,label='(1)',color='red',linestyle='--',linewidth=lnwd)
#
plot(dbl_mod.zs,a.UP,label='(2)',linewidth=lnwd,color='blue',linestyle='-.')
fill_between(dbl_mod.zs,a.UP_min,a.UP_max,alpha=0.5,facecolor='cornflowerblue')
#
plt.title("up ",fontsize=title_s,x=0.44, y=0.8)
plt.ylim([-0.0001,None] )
plt.xlim([0.18,0.7] )
plt.tick_params(axis='both', which='major', labelsize=t_sz)
#
plt.xlabel(r'$z$',size=x_lbl)
plt.ylabel(r'$ z\; D^{\bot (1)}_{1T} (z)$',size=y_lbl)
#
plt.text(0.3, 0.006 , r'(a)', fontsize=45)
plt.text(0.47, 0.004 , r'$Q=10.58$ GeV', fontsize=40)

plt.legend(loc='upper right', fontsize=fnt_s,frameon=False)
#savefig('firstmom_up.png')
#savefig('firstmom_up12.pdf')
```
