```python
import pandas as pd
from numpy import*
```

```python
lp_gauss1 = pd.read_csv('new_bands_/lk_point_gauss_211.2.csv')
lp_gauss2 = pd.read_csv('new_bands_/lk_point_gauss_345.2.csv')
lp_gauss3 = pd.read_csv('new_bands_/lk_point_gauss_3673.2.csv')
```

```python
lp_gauss3
```

```python
lp_tot = lp_gauss1
lp_tot = lp_tot.drop('mins',1)
lp_tot = lp_tot.drop('maxx',1)
lp_tot = lp_tot.drop('valid_min',1)
lp_tot = lp_tot.drop('valid_max',1)
lp_tot = lp_tot.drop('fit_line',1)
```

```python
lp_tot
```

```python
lp_tot['valid']=lp_gauss1['mins']<lp_gauss2['mins']
```

```python
mn = zeros(len(lp_tot.valid))
i=0
for vld,min1,min2 in zip(lp_tot.valid,lp_gauss1.mins,lp_gauss2.mins):
    if vld == True: mn[i] = min1
    elif vld == False: mn[i] = min2

    i+=1 
lp_tot['mins'] = mn    
```

```python
lp_tot
```

```python
lp_tot=lp_tot.drop('valid',1)
lp_tot['valid'] = lp_tot['mins']<lp_gauss3['mins']
```

```python
lp_tot
```

```python
mn = zeros(len(lp_tot.valid))
i=0
for vld,min1,min2 in zip(lp_tot.valid,lp_tot.mins,lp_gauss3.mins):
    if vld == True: mn[i] = min1
    elif vld == False: mn[i] = min2

    i+=1   
lp_tot['mins'] = mn      
```

```python
lp_tot
```

```python
lp_tot=lp_tot.drop('valid',1)
```

```python
lp_tot['valid']=lp_gauss1.maxx>lp_gauss2.maxx 
```

```python
mn = zeros(len(lp_tot.valid))
i=0
for vld,max1,max2 in zip(lp_tot.valid,lp_gauss1.maxx,lp_gauss2.maxx):
    if vld == True: mn[i] = max1
    elif vld == False: mn[i] = max2

    i+=1 
lp_tot['maxx'] = mn    
```

```python
lp_tot=lp_tot.drop('valid',1)
```

```python
lp_tot['valid']=lp_tot.maxx>lp_gauss3.maxx 
```

```python
mn = zeros(len(lp_tot.valid))
i=0
for vld,max1,max2 in zip(lp_tot.valid,lp_tot.maxx,lp_gauss3.maxx):
    if vld == True: mn[i] = max1
    elif vld == False: mn[i] = max2

    i+=1 
lp_tot['maxx'] = mn    
```

```python
lp_tot
```

```python
lp_tot=lp_tot.drop('valid',1)
```

```python
lp_tot
```

```python
lp_tot['valid_min']=lp_tot['conv']>lp_tot['mins']
lp_tot['valid_max']=lp_tot['conv']<lp_tot['maxx']
```
