```python
from first_moments import*
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import*
from matplotlib.pyplot import*
```

```python
dbl_mod=pd.read_csv('first_moments_double_model.csv')
a =pd.read_csv('first_moments_2h.csv')
```

```python
plot(dbl_mod.zs,dbl_mod.zs*a.UP,label='2h')
plot(dbl_mod.zs,dbl_mod.zs*dbl_mod.UP,label='dbl')
fill_between(dbl_mod.zs,dbl_mod.zs*dbl_mod.UP_min,dbl_mod.zs*dbl_mod.UP_max,alpha=0.5)
fill_between(dbl_mod.zs,dbl_mod.zs*a.UP_min,dbl_mod.zs*a.UP_max,alpha=0.5)
legend()
```

```python
plot(dbl_mod.zs,dbl_mod.zs*a.DO,label='2h')
plot(dbl_mod.zs,dbl_mod.zs*dbl_mod.DO,label='dbl')
fill_between(dbl_mod.zs,dbl_mod.zs*dbl_mod.DO_min,dbl_mod.zs*dbl_mod.DO_max,alpha=0.5)
fill_between(dbl_mod.zs,dbl_mod.zs*a.DO_min,dbl_mod.zs*a.DO_max,alpha=0.5)

legend()
```

```python
plot(dbl_mod.zs,dbl_mod.zs*a.ST,label='2h')
plot(dbl_mod.zs,dbl_mod.zs*dbl_mod.ST,label='dbl')
fill_between(dbl_mod.zs,dbl_mod.zs*dbl_mod.ST_min,dbl_mod.zs*dbl_mod.ST_max,alpha=0.5)
fill_between(dbl_mod.zs,dbl_mod.zs*a.ST_min,dbl_mod.zs*a.ST_max,alpha=0.5)

legend()
```
