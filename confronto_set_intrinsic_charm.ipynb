```python
# -*- coding: utf-8 -*-
from pylab import*
import matplotlib.pyplot as plt
from numpy import*
import numpy as np
from matplotlib.pyplot import*
import lhapdf
```

```python
pdf_set1=lhapdf.getPDFSet("CT14nnloIC")
pdf_set2=lhapdf.getPDFSet("NNPDF40_nnlo_as_01180_1000")
```

```python
IC=1
set1=pdf_set1.mkPDF(IC)
set2=pdf_set2.mkPDF(0)
```

```python
xbs2=np.linspace(0.0001,0.9,1000)
q=1.51

pdf_CT=zeros(len(xbs2))
pdf_NN=zeros(len(xbs2))
i=0
for xb2 in xbs2: 
    pdf_CT[i]=set1.xfxQ(4,xb2,q)
    pdf_NN[i]=set2.xfxQ(4,xb2,q)
    i+=1
```

```python
plot(xbs2,pdf_CT,label='CT14')
xscale('log')
plot(xbs2,pdf_NN,label='NNpdf')
legend()
show()
```

```python
set0=pdf_set1.mkPDF(0)
set1=pdf_set1.mkPDF(1)
set2=pdf_set1.mkPDF(2)
set3=pdf_set1.mkPDF(3)
set5=pdf_set1.mkPDF(5)
```

```python
xbs2=np.linspace(0.0001,0.9,1000)
q=1.51

pdf_0=zeros(len(xbs2))
pdf_1=zeros(len(xbs2))
pdf_2=zeros(len(xbs2))
pdf_3=zeros(len(xbs2))
pdf_5=zeros(len(xbs2))

i=0
for xb2 in xbs2: 
    pdf_0[i]=set0.xfxQ(4,xb2,q)
    pdf_1[i]=set1.xfxQ(4,xb2,q)
    pdf_2[i]=set2.xfxQ(4,xb2,q)
    pdf_3[i]=set3.xfxQ(4,xb2,q)
    pdf_5[i]=set5.xfxQ(4,xb2,q)

    i+=1
```

```python
plot(xbs2,pdf_0,label='CT14_0')
plot(xbs2,pdf_1,label='CT14_1')
plot(xbs2,pdf_2,label='CT14_2')
plot(xbs2,pdf_3,label='CT14_3')
plot(xbs2,pdf_5,label='CT14_5')

xscale('log')
legend()
show()
```
