```python
sep=28.6
IC=0
pdf_name='CT14IC'
fnc = cr_sec(sep,IC,pdf_name)
```

```python
had1='lbd_b'
had2='proton'
had22='neutron'
had222='deuterium'
z1=0.25
xb2=0.45
q=10.58
y=0.5
fnc.cross_sec2(had1,had2,z1,xb2,q,y)
```

```python
den=np.array([])
den2=np.array([])
den22=np.array([])
xbs=linspace(0.15,0.6,50)
for xb in xbs:
    den=np.append(den,fnc.cross_sec2(had1,had2,z1,xb,q,y))
    den2=np.append(den2,fnc.cross_sec2(had1,had22,z1,xb,q,y))
    den22=np.append(den22,fnc.cross_sec2(had1,had222,z1,xb,q,y))
```

```python
plot(xbs,den,label='proton')
plot(xbs,den2,label='neutron')
plot(xbs,den22,label='deuterium')
title(r'z1 ='+str(z1),size=15)
yscale('log')
xlabel(r'$x_B$')
legend()
```
