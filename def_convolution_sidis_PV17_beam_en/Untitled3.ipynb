```python
# -*- coding: utf-8 -*-
from pylab import*
import matplotlib.pyplot as plt
from numpy import*
import numpy as np
from matplotlib.pyplot import*
from set_FF.FF_akk import akk_ff
from set_FF.FF_dass import dass_ff
from set_FF.FF_dass07 import dass07_ff
from set_FF.FF_dass14 import dass14_ff
import lhapdf
```

```python
pdf_set2=lhapdf.getPDFSet("NNPDF40_nnlo_as_01180_1000")
pdf_set21=lhapdf.getPDFSet("NNPDF40_nnlo_pch_as_01180")
pdf_set1=lhapdf.getPDFSet("CT14nnloIC")
pdf_set4=lhapdf.getPDFSet("CT14nlo")
```

```python
set1=pdf_set2.mkPDF(0)
set2=pdf_set21.mkPDF(0)
set3=pdf_set1.mkPDF(2)
set4=pdf_set4.mkPDF(0)
```

```python
q=10 #gev
xb2=0.25
pdf_charm=np.array([])
pdf_charm_ct14IC=np.array([])
pdf_pert_charm=np.array([])
pdf_charm_ct14=np.array([])
xbs=np.linspace(0.01,0.9,100)
```

```python
for xb2 in xbs:
    
    pdf_charm=np.append(pdf_charm,set1.xfxQ(4,xb2,q))
    pdf_pert_charm=np.append(pdf_pert_charm,set2.xfxQ(4,xb2,q))
    pdf_charm_ct14IC=np.append(pdf_charm_ct14IC,set3.xfxQ(4,xb2,q))
    pdf_charm_ct14=np.append(pdf_charm_ct14,set4.xfxQ(4,xb2,q))
    
    #print(set1.xfxQ(4,xb2,q))
```

```python
plot(xbs,pdf_pert_charm,label='NNPDF_pert_C')
plot(xbs,pdf_charm,label='NNPDF_IC')
plot(xbs,pdf_charm_ct14IC,label='ct14_IC')
plot(xbs,pdf_charm_ct14,label='ct14_no_C')
yscale('log')
grid()
legend()
```
